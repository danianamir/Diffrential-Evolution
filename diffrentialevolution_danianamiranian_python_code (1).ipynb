```python
import numpy as np
import random
```

```python
bound1=-100
bound2=100
def sphere(x):
    y=(x[0]**2)-(10*x[0])
    return y
```

```python
N=5
D=1
F=0.5
CR=0.7
Iteration=40
```

```python
def initialize_parent():
    parent = np.array([])
    for i in range(N):
        x=np.array([])
        for j in range(D):
            x = np.append(x,random.uniform(bound1,bound2))
        parent= np.concatenate((parent,x))
    parent= np.split(parent,N)
    return parent
```

```python
def choose_mutated(parent):
    mutated=np.array([])
    for i in range(N):
        z=np.array(parent)
        z=np.delete(z,0,axis=0)
        r=random.randint(0,len(z)-1)
        a=z[r]
        z=np.delete(z,r,axis=0)
        r=random.randint(0,len(z)-1)
        b=z[r]
        z=np.delete(z,r,axis=0)
        r=random.randint(0,len(z)-1)
        c=z[r]
        z=np.delete(z,r,axis=0)
        for j in range(D):
            mutated=np.append(mutated,a[j]+F*(b[j]-c[j]))
    mutated= np.split(mutated,N)
    print("parent is :")
    print(parent)
    print("___________________________________________________________________________")
    print("mutated is")
    print(mutated)
    return mutated
```

```python
def check_bound(mutated,parent):
    result=np.array([])
    for i in range(N):
        onboundry=True
        for j in range(D):
                if(not(bound1<=mutated[i][j] and mutated[i][j]<=bound2)):
                    onboundry=False
        if(onboundry):
            result=np.append(result,mutated[i])
        else:
            mutated[i]=np.clip(mutated[i],bound1,bound2)
            result=np.append(result,mutated[i])
    result=np.split(result,N)
    return result
```

```python
def crossover(result,parent):
    cross_points = np.random.rand(D) < CR
    trial=np.array([])
    for i in range(N):
        x= np.where(cross_points, result[i],parent[i])
        trial=np.append(trial,x)
    trial=np.split(trial,N)
    print("___________________________________________________________________________")
    print("trial is:")
    print(trial)
    return trial
```

```python
def selection(trial,parent):
    for i in range(N):
        if(sphere(trial[i])<sphere(parent[i])):
            parent[i]=trial[i]
    print("___________________________________________________________________________")
    print("sphere function amount for 5 population is :")
    for i in range(N):
        print(sphere(parent[i]))
    print("ooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooooo")
    return parent
```

```python
parent=initialize_parent()
for o in range(Iteration):
    mutated=choose_mutated(parent)
    result=check_bound(mutated,parent)
    trial=crossover(result,parent)
    parent= selection(trial,parent)
```
